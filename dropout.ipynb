# Dropout Improvement Technique

This notebook analyzes the impact of adding **dropout layers** to the model architecture. It includes:
- A diagram of the model architecture highlighting the dropout layers.
- Validation loss curves comparing models with and without dropout.

```python
# Import necessary libraries
import pickle
import matplotlib.pyplot as plt

# Paths to the history files
history_without_dropout_path = "ATMEGA_AES_v1/ATM_AES_v1_fixed_key/ASCAD_data/ASCAD_trained_models/my_cnn_best_no_bn_desync0_epochs5_batchsize200_history.pkl"
history_with_dropout_path = "ATMEGA_AES_v1/ATM_AES_v1_fixed_key/ASCAD_data/ASCAD_trained_models/my_cnn_best_dropout_desync0_epochs5_batchsize200_history.pkl"

# Load the history files
with open(history_without_dropout_path, "rb") as f:
    history_without_dropout = pickle.load(f)

with open(history_with_dropout_path, "rb") as f:
    history_with_dropout = pickle.load(f)
```

```python
# Extract validation loss
epochs = range(1, len(history_without_dropout['val_loss']) + 1)
val_loss_without_dropout = history_without_dropout['val_loss']
val_loss_with_dropout = history_with_dropout['val_loss']
```

```python
# Plot validation loss curves
plt.figure(figsize=(10, 6))
plt.plot(epochs, val_loss_without_dropout, label='Validation Loss (Without Dropout)', marker='o')
plt.plot(epochs, val_loss_with_dropout, label='Validation Loss (With Dropout)', marker='o')
plt.title('Validation Loss Comparison')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid()
plt.show()
```
